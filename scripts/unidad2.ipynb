## Ejemplos If Condicional

### Ejemplo 1

```python
edad = int(input("Cual es la edad: "))

if edad >= 18:
    print("Eres mayor de edad")
else:
    print("Eres menor de edad")
```

### Ejemplo 2

```python
calificacion = float(input("Ingresa la calificación: "))

if calificacion >= 9:
    print("Excelente")
elif calificacion >= 6:
    print("Aprobado")
else:
    print("Reprobado")
```

### Ejemplo 3

```python
anio = int(input("Ingresa un año: "))

if (anio % 4 == 0 and anio % 100 != 0) or (anio % 400 == 0):
    print("Es un año bisiesto")
else:
    print("No es un año bisiesto")
```

### Ejemplo 4

```python
a = float(input("Ingresa el primer número: "))
b = float(input("Ingresa el segundo número: "))
opcion = int(input("Elige una opción (1-4): "))

if opcion == 1:
    print("Resultado:", a + b)
elif opcion == 2:
    print("Resultado:", a - b)
elif opcion == 3:
    print("Resultado:", a * b)
elif opcion == 4:
    if b != 0:
        print("Resultado:", a / b)
    else:
        print("Error: división entre cero")
else:
    print("Opción inválida")
```

## Ejemplos For

### Ejemplo 1

```python
n = int(input("¿Cuántas veces deseas mostrar el mensaje?: "))

for i in range(n):
    print("Hola Python")
```

### Ejemplo 2

```python
cantidad = int(input("¿Cuántos números vas a ingresar?: "))
suma = 0

for i in range(cantidad):
    numero = float(input(f"Ingrese el número {i + 1}: "))
    suma += numero

print("La suma total es:", suma)
```

### Ejemplo 3

```python
cantidad = int(input("¿Cuántos números deseas ingresar?: "))
contador_pares = 0

for i in range(cantidad):
    numero = int(input(f"Ingrese el número {i + 1}: "))
    
    if numero % 2 == 0:
        contador_pares += 1

print("Cantidad de números pares:", contador_pares)
```

## Ejemplos While

### Ejemplo 1

```python
while True:
    try:
        numero = float(input("Ingresa un número positivo: "))
        if numero > 0:
            break
        else:
            print("El número debe ser mayor que cero")
    except ValueError:
        print("Entrada inválida, intenta de nuevo")

print("Número válido:", numero)
```

### Ejemplo 2

```python
opcion = 0

while opcion != 3:
    print("\nMenú")
    print("1. Saludar")
    print("2. Mostrar mensaje")
    print("3. Salir")

    opcion = int(input("Elige una opción: "))

    if opcion == 1:
        print("¡Hola! Bienvenido")
    elif opcion == 2:
        print("Python es genial")
    elif opcion == 3:
        print("Saliendo del programa...")
    else:
        print("Opción inválida")
```

### Ejemplo 3

```python
while True:
    try:
        numero = int(input("Ingresa un número entero positivo: "))
        if numero >= 0:
            break
        else:
            print("El número debe ser positivo")
    except ValueError:
        print("Entrada inválida")

while numero >= 0:
    print(numero)
    numero -= 1
```

## Ejemplo Clases

```python
class Alumno:
    def __init__(self, nombre, edad, promedio): # es un método especial de las clases en Python. Su función es inicializar (dar valores         
                                                # iniciales) a los atributos del objeto cuando este se crea.
        self.nombre = nombre  #self es una referencia al objeto actual. Permite que cada objeto tenga sus propios datos.
        self.edad = edad
        self.promedio = promedio

    def mostrar_datos(self):
        print("Nombre:", self.nombre)
        print("Edad:", self.edad)
        print("Promedio:", self.promedio)

    def esta_aprobado(self):
        return self.promedio >= 6


# Programa principal
alumno1 = Alumno("Ana", 20, 8.5)
alumno1.mostrar_datos()

if alumno1.esta_aprobado():
    print("Aprobado")
else:
    print("Reprobado")
```

## Ejemplo Excepciones

### Ejemplo 1

```python
while True:
    try:
        edad = int(input("Ingresa tu edad: "))

        if edad < 0 or edad > 120:
            print("La edad debe estar entre 0 y 120")
        else:
            break

    except ValueError:
        print("Error: debes ingresar un número entero")

print("Edad registrada:", edad)
```

### Ejemplo 2

```python
try:
    num1 = float(input("Ingresa el primer número: "))
    num2 = float(input("Ingresa el segundo número: "))

    resultado = num1 / num2
    print("Resultado:", resultado)

except ZeroDivisionError:
    print("Error: no se puede dividir entre cero")

except ValueError:
    print("Error: debes ingresar números válidos")

else:
    print("Operación realizada correctamente")

finally:
    print("Fin del programa")
```
